```python
from google.colab import files
import pandas as pd
import zipfile

# This will show an upload button
uploaded = files.upload()

# Get the uploaded filename
for fn in uploaded.keys():
    filename = fn

# If it's a zip file, extract it
if filename.endswith(".zip"):
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall(".")
    # Assuming deliveries.csv is inside
    df = pd.read_csv("deliveries.csv")
else:
    df = pd.read_csv(filename)

# Show first 5 rows
df.head()
```

```python
# Check number of rows and columns
print("Shape of dataset:", df.shape)

# Rows count
print("Number of rows:", df.shape[0])

# Columns count
print("Number of columns:", df.shape[1])
```

```python
# Total number of matches
total_matches = df['match_id'].nunique()

# Number of wide balls
num_wides = df['wide_runs'].astype(bool).sum()
avg_wides_per_match = num_wides / total_matches

# Number of no-balls
num_noballs = df['noball_runs'].astype(bool).sum()
avg_noballs_per_match = num_noballs / total_matches

# Number of 6s hit
num_sixes = (df['batsman_runs'] == 6).sum()
avg_sixes_per_match = num_sixes / total_matches

# Number of 4s hit
num_fours = (df['batsman_runs'] == 4).sum()
avg_fours_per_match = num_fours / total_matches

print("ðŸ‘‰ Averages per Match:")
print(f"Wide Balls: {avg_wides_per_match:.2f}")
print(f"No-Balls: {avg_noballs_per_match:.2f}")
print(f"Sixes: {avg_sixes_per_match:.2f}")
print(f"Fours: {avg_fours_per_match:.2f}")
```

```python
import pandas as pd

# 1. Top 10 Run Scorers
top_runs = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print("ðŸ Top 10 Run Scorers:\n", top_runs)

# 2. Top 10 Boundary Hitters (4s and 6s)
fours = df[df['batsman_runs'] == 4].groupby('batsman')['batsman_runs'].count()
sixes = df[df['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].count()

boundaries = pd.DataFrame({
    '4s': fours,
    '6s': sixes,
    'Total Boundaries': fours.add(sixes, fill_value=0)
}).fillna(0).astype(int).sort_values('Total Boundaries', ascending=False).head(10)

print("\nðŸ”¥ Top 10 Boundary Hitters:\n", boundaries)

# 3. Strike Rate (Runs per 100 balls) - Top 10 by runs
balls_faced = df.groupby('batsman')['ball'].count()
runs_scored = df.groupby('batsman')['batsman_runs'].sum()
strike_rate = (runs_scored / balls_faced * 100).sort_values(ascending=False)

top_strike_rate = strike_rate.loc[runs_scored.nlargest(10).index]
print("\nâš¡ Top 10 Batsmen by Strike Rate (among top run scorers):\n", top_strike_rate)
```

```python
# 1. Top Wicket Takers
# A dismissal is valid only if 'player_dismissed' is not null
wickets = df[df['player_dismissed'].notnull()].groupby('bowler')['player_dismissed'].count()
top_wickets = wickets.sort_values(ascending=False).head(10)
print("ðŸŽ¯ Top 10 Wicket Takers:\n", top_wickets)

# 2. Best Economy Rate (min 200 balls bowled)
balls_bowled = df[df['wide_runs']==0][df['noball_runs']==0].groupby('bowler')['ball'].count()
runs_conceded = df.groupby('bowler')['total_runs'].sum()

economy = (runs_conceded / (balls_bowled/6)).dropna()
economy = economy[balls_bowled > 200].sort_values().head(10)
print("\nðŸ’° Best Economy Rates (min 200 balls bowled):\n", economy)

# 3. Dot Ball Specialists
dot_balls = df[df['total_runs']==0].groupby('bowler')['ball'].count()
top_dot_balls = dot_balls.sort_values(ascending=False).head(10)
print("\nðŸ›‘ Top 10 Dot Ball Specialists:\n", top_dot_balls)

```

```python
# Runs scored per match & inning
team_totals = df.groupby(['match_id','inning','batting_team'])['total_runs'].sum().reset_index()

# Top 10 highest totals
top_totals = team_totals.sort_values('total_runs', ascending=False).head(10)
print("ðŸ”¥ Top 10 Team Totals:\n", top_totals)

avg_runs_per_over = df.groupby('over')['total_runs'].mean() * 6
print("\nðŸ“Š Average Runs per Over:\n", avg_runs_per_over)

extras_by_team = df.groupby('bowling_team')['extra_runs'].sum().sort_values(ascending=False)
print("\nâš ï¸ Extras Conceded by Teams:\n", extras_by_team)

# Runs scored
runs_scored = df.groupby('batting_team')['total_runs'].sum()

# Runs conceded
runs_conceded = df.groupby('bowling_team')['total_runs'].sum()

team_strength = pd.DataFrame({'Runs Scored': runs_scored, 'Runs Conceded': runs_conceded})
print("\nðŸ’ª Team Batting vs Bowling Strength:\n", team_strength)
```

```python
# Powerplay = overs 1â€“6, exclude super overs just in case
pp = df[(df['is_super_over'] == 0) & (df['over'].between(1, 6))]

# --- Option A: All dismissals (includes run-outs) ---
pp_summary_all = pp.groupby(['match_id', 'inning']).agg(
    runs=('total_runs', 'sum'),
    wickets=('player_dismissed', lambda s: s.notna().sum())
).reset_index()

print("Per-innings Powerplay (All dismissals)")
print("Avg runs :", pp_summary_all['runs'].mean())
print("Avg wkts :", pp_summary_all['wickets'].mean())


# Middle overs
middle = df[(df['over'] >= 7) & (df['over'] <= 15)]
mid_runs = middle.groupby('match_id')['total_runs'].sum().mean()
mid_wkts = middle['player_dismissed'].notnull().sum() / df['match_id'].nunique()

# Death overs
death = df[(df['over'] >= 16) & (df['over'] <= 20)]
death_runs = death.groupby('match_id')['total_runs'].sum().mean()
death_wkts = death['player_dismissed'].notnull().sum() / df['match_id'].nunique()

print(f"ðŸŒ€ Middle Overs: Avg Runs = {mid_runs:.2f}, Avg Wickets = {mid_wkts:.2f}")
print(f"ðŸ’€ Death Overs: Avg Runs = {death_runs:.2f}, Avg Wickets = {death_wkts:.2f}")

super_over = df[df['is_super_over'] == 1]
if not super_over.empty:
    so_runs = super_over.groupby('match_id')['total_runs'].sum()
    print("\nðŸ”¥ Super Over Runs by Match:\n", so_runs)
else:
    print("\nNo Super Overs found in this dataset.")

# Filter only death overs (16-20)
death = df[df['over'] >= 16]

# Group by batsman: total runs & balls faced
death_batsmen = (
    death.groupby('batsman')
    .agg(
        runs=('batsman_runs', 'sum'),
        balls=('ball', 'count')
    )
)

# Calculate Strike Rate (SR = Runs / Balls * 100)
death_batsmen['SR'] = (death_batsmen['runs'] / death_batsmen['balls']) * 100

# Sort by runs (descending) and take top 10
death_batsmen = death_batsmen.sort_values(by='runs', ascending=False).head(10)

print("\nðŸ’¥ Top 10 Death Overs Batsmen (Runs + Strike Rate):\n")
print(death_batsmen)

# Bowlers in death overs (most wickets)
death_wickets = death[death['player_dismissed'].notnull()].groupby('bowler')['player_dismissed'].count()
death_bowlers = death_wickets.sort_values(ascending=False).head(10)
print("\nðŸ§¨ Top 10 Death Overs Bowlers:\n", death_bowlers)
```

```python
import pandas as pd

# Total matches
total_matches = df['match_id'].nunique()

# Insights
num_wides = df['wide_runs'].astype(bool).sum()
num_noballs = df['noball_runs'].astype(bool).sum()
num_sixes = (df['batsman_runs'] == 6).sum()
num_fours = (df['batsman_runs'] == 4).sum()

# Averages
avg_wides = num_wides / total_matches
avg_noballs = num_noballs / total_matches
avg_sixes = num_sixes / total_matches
avg_fours = num_fours / total_matches

# Create summary dataframe
summary = pd.DataFrame({
    "Metric": ["Wide Balls", "No Balls", "Sixes", "Fours"],
    "Total": [num_wides, num_noballs, num_sixes, num_fours],
    "Average per Match": [avg_wides, avg_noballs, avg_sixes, avg_fours]
})

# Save to Excel
summary.to_excel("cricket_summary.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Runs per Over
avg_runs_per_over = df.groupby('over')['total_runs'].mean()

plt.figure(figsize=(10,6))
plt.plot(avg_runs_per_over.index, avg_runs_per_over.values, marker='o')
plt.title("ðŸ“Š Average Runs per Over", fontsize=14)
plt.xlabel("Over")
plt.ylabel("Average Runs")
plt.grid(True)
plt.show()

# 2. Wickets per Over
wickets_per_over = df[df['player_dismissed'].notna()].groupby('over')['player_dismissed'].count()

plt.figure(figsize=(10,6))
plt.bar(wickets_per_over.index, wickets_per_over.values, color="crimson")
plt.title("ðŸ’€ Wickets per Over", fontsize=14)
plt.xlabel("Over")
plt.ylabel("Number of Wickets")
plt.show()

# 3. Top 10 Batsmen in Death Overs
death = df[df['over'] >= 16]
death_batsmen = death.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
death_batsmen.plot(kind="bar", color="orange")
plt.title("ðŸ’¥ Top 10 Death Overs Batsmen", fontsize=14)
plt.xlabel("Batsman")
plt.ylabel("Runs in Death Overs")
plt.xticks(rotation=45)
plt.show()

# 4. Distribution of Sixes & Fours
six_count = (df['batsman_runs'] == 6).sum()
four_count = (df['batsman_runs'] == 4).sum()

plt.figure(figsize=(6,6))
plt.pie([six_count, four_count], labels=["Sixes", "Fours"], autopct='%1.1f%%', colors=["gold", "skyblue"], startangle=90)
plt.title("ðŸŽ¯ Sixes vs Fours", fontsize=14)
plt.show()

# 5. Extras (Wides & No Balls)
num_wides = df['wide_runs'].astype(bool).sum()
num_noballs = df['noball_runs'].astype(bool).sum()

plt.figure(figsize=(6,6))
plt.bar(["Wides", "No Balls"], [num_wides, num_noballs], color=["purple", "teal"])
plt.title("âš ï¸ Extras (Wides & No Balls)", fontsize=14)
plt.ylabel("Count")
plt.show()

# 6. Powerplay Runs & Wickets
powerplay = df[df['over'] <= 6]
pp_runs = powerplay['total_runs'].sum()
pp_wickets = powerplay['player_dismissed'].notna().sum()

plt.figure(figsize=(6,6))
plt.bar(["Runs", "Wickets"], [pp_runs, pp_wickets], color=["green", "red"])
plt.title("ðŸš€ Powerplay Impact", fontsize=14)
plt.ylabel("Total")
plt.show()
```
